```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataframe= pd.read_csv("Zomato data .csv")
print(dataframe)
```

```python
dataframe
```

# **Conversion of Data Type of column - rate**

```python
def handleRate(value):
  value = str(value).split('/')
  value= value[0]
  return float(value)

dataframe ['rate'] = dataframe['rate'].apply(handleRate)
print(dataframe)
```

# **check the missing values in the data**

```python
dataframe.info()
```

# **Finding Types of restaurant**

```python
dataframe.head()
# head use to get the first five data from csv file
```

```python
sns.countplot(x=dataframe['listed_in(type)'], palette='Set2')
plt.xlabel("type of restaurant")
plt.ylabel("no. of restaurants")
plt.title("Majority")
plt.show()
```

# **Conclusion of 1. is = Dining is majority in restaurants customers order from**

```python
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes':grouped_data})
plt.plot(result, c= "yellow", marker="o")
plt.xlabel("type of restaurant", c= "green", size=20)
plt.ylabel("no. of votes", c="red", size=20)
plt.title("Votes", c="blue", size=30)

```

# **Conclusion of 2.= Dinning restaurants has maximum votes**

```python
plt.hist(dataframe['rate'], bins=6) #bins are used for dividing the columns in parts
plt.title("rating of restaurants")
plt.show()
```

# **Conclusion of 4.= The majority of couples prefer restaurants with an appropriate cost of 300 rupees**


```python
plt.figure(figsize=(6,6))
sns.boxplot(x='online_order', y='rate', data=dataframe, palette='Set2')
plt.xlabel("online order")
plt.ylabel("rating")
plt.show()
```

# **Conclusion of 5.= Offline order recieved lower rating in comparision to online order**

```python
pivot_table = dataframe.pivot_table(index='listed_in(type)', columns='online_order', aggfunc='size', fill_value=0)
sns.heatmap(pivot_table, annot=True, cmap='YlGnBu')
plt.xlabel("online order")
plt.title("heatmap")
plt.ylabel("Listed In (Type)")
plt.show()
```

# **Dining restaurants primarily accept offline orders, whereas cafes primarily receive online orders.This suggests that clients prefer to place orders in person at restaurants, but prefere online ordering at cafes.**
